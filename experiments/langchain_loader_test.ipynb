```python
loader = PyPDFLoader("test.pdf")
docs = loader.load_and_split()
docs
```

```python
import os
import time
from dotenv import load_dotenv
from pinecone import Pinecone, ServerlessSpec
from langchain_mistralai.embeddings import MistralAIEmbeddings
```

```python
docs_portions = docs
```

```python
load_dotenv()
```

```python
pinecone_api_key = os.getenv("PINECONE_API_KEY")
if not pinecone_api_key:
    raise ValueError("PINECONE_API_KEY is not set")

pc = Pinecone(api_key=pinecone_api_key)
```

```python
index_name = "langchain-test-index" 

existing_indexes = [index_info["name"] for index_info in pc.list_indexes()]
```

```python
if index_name not in existing_indexes:
    pc.create_index(
        name=index_name,
        dimension=1024,
        metric="cosine",
        spec=ServerlessSpec(cloud="aws", region="us-east-1"),
    )
    while not pc.describe_index(index_name).status["ready"]:
        time.sleep(1)

index = pc.Index(index_name)
```

```python
print(index)
```

```python
mistral_api_key = os.getenv("MISTRAL_API_KEY")
if not mistral_api_key:
    raise ValueError("MISTRAL_API_KEY is not set")

embeddings = MistralAIEmbeddings(model="mistral-embed", api_key=mistral_api_key)
```

```python
from langchain_pinecone import PineconeVectorStore
vector_store = PineconeVectorStore(index=index, embedding=embeddings, namespace="test")
```

```python
from uuid import uuid4
docs_portions = docs[10:]
uuids = [str(uuid4()) for _ in range(len(docs_portions))]
```

```python
len(uuids)
```

```python
vector_store.add_documents(documents=docs_portions, ids=uuids, namespace="test")
```

```python
results = vector_store.similarity_search("Christine Stephen")
```

```python
results
```

```python
from langchain_community.tools import tool
from langchain_mistralai import ChatMistralAI

@tool
def search_vector_store(query: str):
    """
        Search the database for the given query
    """
    return vector_store.similarity_search(query)

@tool
def addition(a: int, b: int):
    """
        Add two numbers
    """
    return a + b
```

```python
llm = ChatMistralAI(model="mistral-large-latest")
```

```python
from langgraph.prebuilt import create_react_agent

agent = create_react_agent(llm, [search_vector_store, addition])
```

```python
from langchain_core.messages import HumanMessage

result = agent.invoke({"messages": [HumanMessage(content="Add 2 and 2")]})
result
```
